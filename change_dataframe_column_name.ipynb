```python
#changing dataframe column name

import pandas as pd

df =pd.DataFrame({'col one':[100,200],'col two':[300,400]})
df

#rename korar first way

df = df.rename({'col one':'Colum_One'},axis='columns')
df

#second column

df= df.rename({'col two':'Colum_Two'},axis = 'columns')
df

```

```python
#2nd way of renaming the column

import pandas as pd

df =pd.DataFrame({'col one':[100,200],'col two':[300,400]})
df

df.columns = ['Column_1','Column_2']


df
```

```python
#3rd way of renaming of a column


import pandas as pd

df =pd.DataFrame({'col one':[100,200],'col two':[300,400]})
df

df.columns=df.columns.str.replace(' ','_')
df
```
