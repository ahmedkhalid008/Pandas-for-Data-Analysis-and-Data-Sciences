```python
import pandas as pd

climate_data ={
    'Day' :['01.01.2023','02.01.2023','03.01.2023'],
    'Temperature':[31,27,30],
    'Windspeed' :[3,6,9],
    'event':['Sunny','Rainy','Cloudy']
}

df = pd.DataFrame(climate_data)
df

#single dataframe update using single condition condition

df.at[2,'Temperature']=40
df
df.at[1,'Temperature']= 50
df


#multiple dataframe update using multiple condition

df.loc[df['Windspeed']>3 ,['event']]='Sunny'
df
df.loc[df['Windspeed']==3,['event']]='Cloudy'
df
```
