```python
#Creating a Table using dict. data

import pandas as pd
student_data= {
    'Name' : ['Khalid','Rahim','Karim'],
     'ID'  : [8,9,10],
    'Department':['IT','EEE','CSE'],
    'HomeTown' : ['Gopalganj','Dhaka','Khulna']
}

df= pd.DataFrame(student_data)
#print(df)
df
```
